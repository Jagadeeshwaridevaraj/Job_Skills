```python
from collections import Counter
import pandas as pd

# Convert the skills column to lowercase
jobs_df['Skills'] = jobs_df['Skills'].str.lower()

# Split skills into individual items and flatten the list
skills_list = jobs_df['Skills'].str.split(',').explode().str.strip()

# Count unique skills
skills_count = Counter(skills_list)

# Convert the count to a DataFrame
skills_count_df = pd.DataFrame(list(skills_count.items()), columns=['Skill', 'Count'])

# Sort by count in descending order
skills_count_df = skills_count_df.sort_values(by='Count', ascending=False).reset_index(drop=True)

skills_count_df
```

```python

# Function to remove "data" and keep only the next word
def remove_data_word(skill):
    if pd.isna(skill) or not isinstance(skill, str):  # Handle NaN or non-string values
        return skill
    words = skill.split()
    if words and words[0].lower() == 'data':  # Ensure words list is not empty
        return ' '.join(words[1:]) if len(words) > 1 else ''  # Keep only the next words
    return skill  # Return as is if "data" is not the first word

# Apply function to the 'Skill' column
skills_count_df['Skill'] = skills_count_df['Skill'].apply(remove_data_word)

# Display updated DataFrame
print(skills_count_df)
```
